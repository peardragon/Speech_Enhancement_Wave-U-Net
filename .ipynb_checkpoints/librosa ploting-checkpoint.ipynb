```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
import torchvision
```

```python
wav, _ = librosa.load('./testset/noisy/p232_023.wav', sr=22050, mono=True, res_type='kaiser_fast', offset=0.0, duration=None)
```

```python
D = librosa.stft(wav)  # STFT of y
S_db1 = librosa.amplitude_to_db(np.abs(D), ref=np.max)
# return shape (1+n_fft/2, n_frames)
fig = plt.figure()
librosa.display.specshow(S_db1, y_axis='log', x_axis='time')
plt.colorbar(format="%+2.f dB")
plt.show()
```

```python
fig = plt.figure()
librosa.display.waveplot(wav)
plt.show()
```
